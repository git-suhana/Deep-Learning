```python
import tensorflow as tf
```

```python
from tensorflow.keras.utils import img_to_array,array_to_img,load_img
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
datagen= ImageDataGenerator(rotation_range=40,width_shift_range=0.2,shear_range=0.2,horizontal_flip=True,fill_mode='nearest')
```

```python
img= load_img('dolphin.jpg')
```

```python
x= img_to_array(img)    
```

```python
x.shape
```

```python
x=x.reshape((1,)+x.shape)   
```

```python
x.shape
```

```python
i=0


for batch in datagen.flow(x,batch_size=1,save_to_dir='img1_dir',save_prefix='cl',save_format='jpeg'):
    i+=1
    if i>20:
        break            
```
